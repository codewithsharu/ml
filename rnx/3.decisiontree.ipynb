```python
# Import necessary libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import tree
import matplotlib.pyplot as plt

# Step 1: Load your dataset (decisiontree.csv)
data = pd.read_csv('decisiontree.csv')

# Step 2: Separate the features (X) and target (y)
X = data.iloc[:, :-1]  # All columns except the last one as features
y = data.iloc[:, -1]   # Last column as the target

# Step 3: Split the dataset into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Initialize the Decision Tree Classifier (using entropy for ID3)
clf = DecisionTreeClassifier(criterion='entropy')
clf.fit(X_train, y_train)  # Train the model

# Step 5: Predict the class for a new sample (replace with actual values from your dataset)
# Example values for the new sample; replace these with actual feature values
value1 = 5.0  # Replace with actual value for Feature1
value2 = 3.5  # Replace with actual value for Feature2
value3 = 1.5  # Replace with actual value for Feature3
value4 = 0.2  # Replace with actual value for Feature4

new_sample = [[value1, value2, value3, value4]]  # Replace with actual feature values
prediction = clf.predict(new_sample)
print(f"Predicted class for the new sample {new_sample}: {prediction[0]}")

# Step 6: Visualize the decision tree
plt.figure(figsize=(15, 10))
tree.plot_tree(clf, feature_names=X.columns, class_names=clf.classes_, filled=True)
plt.show()

# Step 7: Evaluate the accuracy on the test set
accuracy = clf.score(X_test, y_test)
print(f"Accuracy on the test set: {accuracy * 100:.2f}%")
```
